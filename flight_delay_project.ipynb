```python
df['ScheduledArrival']=pd.to_datetime(df['ScheduledArrival'])
df['ActualArrival']=pd.to_datetime(df['ActualArrival'])
df['ScheduledDeparture']=pd.to_datetime(df['ScheduledDeparture'])
df['ActualDeparture']=pd.to_datetime(df['ActualDeparture'])
df.info()
```

12. Compare between arrival delays and departure delays?

```python
df['DepartureDelay']=(df['ActualDeparture']-df['ScheduledDeparture']).dt.total_seconds()/60
df['ArrivalDelay']=(df['ActualArrival']-df['ScheduledArrival']).dt.total_seconds()/60
delay_comparison=df[['DepartureDelay','ArrivalDelay']]
print(delay_comparison)
```

```python
plt.xticks(rotation=45)
sns.boxplot(x='DepartureDelay',y='ArrivalDelay',data=df)
plt.title('Comparison of Departure and Arrival Delays')
plt.xlabel('Departure Delay in minutes')
plt.ylabel('Arrival Delay in minutes')
plt.show()
```

*Each box represents the interquartile range (IQR), which is considered as the 50% of data, the line inside the box represents the median arrival delay for that specific departure delay value.*

*For each departure delay (represented on the x axis),the distribution of arrival delays are shown(represented on the y axis).*

*The boxes are relatively aligned- when we observe at the arrival delay, the upper extreme of the boxplot (30 minutes) are constant for each departure delay, the lower extreme of the boxplot (-10 minutes), which is a negative delay, it means the flight has arrived 10 minutes earlier than it's scheduled arrival time.*

*The median arrival delay follows a linear path in correspondence with departure delay. Here 10 minutes is considered as the median delay minute.*

*We can find that the overall distribution of delays proves that almost all flights follows a predictable delay pattern.*
